## Model Oluşturma
**Bu veri bir unsupervised learning dosyası cünkü unlabelled data var ve unsupervised ve çok fazla özelliğin olduğu bu tip verilerle çalışırken ve anomali tespiti yaparken 2007 de yayınlanan isolation forest adlı algoritma oldukça sık kullanılır bu yüzden biz de oldukça iyi bir algoritma olması az ram kullanımı ve verimize uygunluğundan dolayı bu algoritmayı seçtik.**
**Isolation Trees ile daha cok bilgi sahibi olmak isterseniz daha çok bilgiyi bu linkten bulabilirsiniz**
**https://en.wikipedia.org/wiki/Isolation_forest**

```python
#modelimizi kuralım
from sklearn.ensemble import IsolationForest
#contamination valuemizi genelde isolation forestin optimum çalıştığı değer olarak verdik number of estimators 
#değeri hiç hassas olmadığı için genel bir değer verdik detaylı bir analiz olması adına 1000'i uygun gördüm
contamination = 0.17
##üstünde herhangi bir hata olduğu zaman geri dönebilmek için geçici bir veritabanı tasarladık 
data=df.copy()
```

```python
data
```

**Isolation Trees kullanırken tüm verilerimizin numerik değerlere indirgenmesi gerekmektedir bu yüzden object nesnemiz olan tarih özelliğimizi label encoder yardımıyla numerikleştirdik**

```python
for col in data.columns:
    if data[col].dtype=="object":
        le=preprocessing.LabelEncoder()
        le.fit(list(data[col].astype(str).values))
        data[col]= le.transform(list(data[col].astype(str).values))
```

**Verimizi modelimize bağladık**

```python
%%time
model=IsolationForest(contamination=contamination,n_estimators=1000)
model.fit(data)
```

```python
df['iforest']= pd.Series(model.predict(data))
df['iforest']= df['iforest'].map({1:0,-1:1})
print(df['iforest'].value_counts())
```
